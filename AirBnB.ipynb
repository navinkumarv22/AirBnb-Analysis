## Handling Missing Values

### Column with 22.6% missing values

```python
df['review_scores_rating'].fillna(df['review_scores_rating'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

### Columns with negligible missing value percentage

```python
df.dropna(subset=['room_type', 'accommodates', 'bathrooms', 'cancellation_policy', 'cleaning_fee', 'bedrooms', 'beds'], inplace=True)
```

```python
df.isnull().sum()
```

## Univariate Analysis

### Count Plot of Room type distribution

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df, y='room_type', 
              order=df['room_type'].value_counts().index,
             palette='pastel')
plt.title("Room type distribution", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Room Type",fontsize=12)
plt.show()
```

#### Inference:
* Most listings are likely Entire Home/Apt.
* Shared room is available only in small numbers.

### Histogram of Price Distribution

```python
sns.histplot(df['log_price'], bins=30, kde=True)
plt.title('Log Price Distribution')
plt.xlabel('Log Price')
plt.ylabel('Frequency')
plt.show()
```

#### Inference:
* The histogram follows a bell-shaped curve, indicating a normal distribution of log-transformed prices.
* The highest frequency of listings falls around log price 4.5 to 5.5.
* Some listings have log price values above 6 and 7, suggesting high-end or luxury properties.
* Very few listings have a log price below 3, meaning budget listings are rare in the dataset.

### Histogram of Review scores distribution

```python
sns.histplot(df['review_scores_rating'], bins=30, kde=True)
plt.title('Review Scores Distribution')
plt.show()
```

#### Inference:
* We see a left-skewed distribution, indicating that the most of the guests liked the rooms and gave high ratings.
* Rooms with less rating indicate that the customers were dissatified with the rooms.

## Bivariate/Multi Variate Analysis

### Line Chart of Log Price VS Accomodates

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["accommodates"], y=df["log_price"], alpha=0.5)
plt.title("Log Price Based on Number of Accommodates")
plt.xlabel("Accommodates")
plt.ylabel("Log Price")
plt.grid(True)
plt.show()
```

#### Inference:
* This chart shows how pricing changes with the number of people a listing can accommodate.
* It likely follows a positive trend, meaning larger accommodations (with more capacity) tend to have higher prices.
* The trend may not be linear—prices could increase at a decreasing rate after a certain threshold.

### Line chart of Log Price VS Room type

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["room_type"], y=df["log_price"], alpha=0.5)
plt.title("Log Price Across Room Types")
plt.xlabel("Room Type")
plt.ylabel("Log Price")
plt.grid(True)
plt.show()
```

#### Inference:
* Entire homes/apartments have the highest log price, followed by private rooms, and shared rooms have the lowest.
* This suggests that the level of privacy and exclusivity significantly influences price.

### Correlation Matrix of Rooms

```python
df["instant_bookable"] = df["instant_bookable"].map({"t": 1, "f": 0})
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[["log_price", "accommodates", "beds", "review_scores_rating", "instant_bookable"]].corr(), 
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Airbnb Listings")
plt.show()
```

#### Inference:
* Listings that accommodate more guests tend to have higher prices.
* More beds generally lead to higher prices, but this correlation might be weaker than accommodates.
* This suggests that capacity (accommodates) has a stronger influence on pricing than just the number of beds.

### Scatter plot of Log Price VS Review scores

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["review_scores_rating"], y=df["log_price"], alpha=0.5)
plt.title("Log Price vs. Review Scores Rating")
plt.xlabel("Review Scores Rating")
plt.ylabel("Log Price")
plt.grid(True)
plt.show()
```

#### Inference:
* High ratings are common, but they don’t necessarily mean a high price.
* Some expensive properties maintain high ratings, but price alone doesn’t guarantee better reviews.
* The presence of clustered values indicates filling with mean of review score.
